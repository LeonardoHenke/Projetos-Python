```python
import pandas as pd

# dando entrada nas tabelas que serÃ£o usadas
produtos_df = pd.read_csv(r'E:\Programas\Projetos-Python\Pandas\Contoso - Cadastro Produtos.csv', sep=';')
clientes_df = pd.read_csv(r'E:\Programas\Projetos-Python\Pandas\Contoso - Clientes.csv', sep=';')
lojas_df = pd.read_csv(r'E:\Programas\Projetos-Python\Pandas\Contoso - Lojas.csv', sep=';')
vendas2017_df = pd.read_csv(r'E:\Programas\Projetos-Python\Pandas\Contoso - Vendas  - 2017.csv', sep=';')

# filtrando as tabelas conforme as colunas que queremos
clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto', 'Nome da Marca']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]

# mesclando e renomeando colunas
vendas2017_df = vendas2017_df.merge(produtos_df, on='ID Produto')
vendas2017_df = vendas2017_df.merge(clientes_df, on='ID Cliente')
vendas2017_df = vendas2017_df.merge(lojas_df, on='ID Loja').rename(columns={'E-mail' : 'E-mail do Cliente'})
display(vendas2017_df)
```

Transformar a coluna de data da venda em colunas diferentes para mÃªs, dia e ano
    - pd.to_datetima(argumento, format='formato que vc deseja') - argumento Ã© o que vc deseja passar para formato de data, por exemplo a coluna no exercÃ­cio abaixo
    - vendas2017_df['argumento'] - Se nÃ£o existir uma coluna com esse Ã­ndice, ele ira criar uma coluna
    - .dt.argumento - usar day, month ou year como argumento visando o dado desejado

```python
# transformando a coluna data da venda do formato objeto para uma data
vendas2017_df['Data da Venda'] = pd.to_datetime(vendas2017_df['Data da Venda'], format='%d/%m/%Y') # colocar o mesmo formato utilizado na base(no caso, utilizado na coluna em questÃ£o)
#display(vendas2017_df['Data da Venda'])

# criando colunas para o dia, mÃªs e ano
vendas2017_df['Dia da Venda'] = vendas2017_df['Data da Venda'].dt.day
vendas2017_df['Mes da Venda'] = vendas2017_df['Data da Venda'].dt.month
vendas2017_df['Ano da Venda'] = vendas2017_df['Data da Venda'].dt.year
display(vendas2017_df)

# removendo a coluna data da venda
vendas2017_df = vendas2017_df.drop(['Data da Venda'], axis=1)
display(vendas2017_df)
```
