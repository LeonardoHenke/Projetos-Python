```python
import pandas as pd

# dando entrada nas tabelas que serão usadas
produtos_df = pd.read_csv(r'E:\Programas\Projetos-Python\Pandas\Contoso - Cadastro Produtos.csv', sep=';')
clientes_df = pd.read_csv(r'E:\Programas\Projetos-Python\Pandas\Contoso - Clientes.csv', sep=';')
lojas_df = pd.read_csv(r'E:\Programas\Projetos-Python\Pandas\Contoso - Lojas.csv', sep=';')
novas_vendas2017_df = pd.read_csv(r'E:\Programas\Projetos-Python\Pandas\Contoso - Vendas  - 2017.csv', sep=';')

# filtrando as tabelas conforme as colunas que queremos
clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto', 'Nome da Marca']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]

# mesclando e renomeando colunas
novas_vendas2017_df = novas_vendas2017_df.merge(produtos_df, on='ID Produto')
novas_vendas2017_df = novas_vendas2017_df.merge(clientes_df, on='ID Cliente')
novas_vendas2017_df = novas_vendas2017_df.merge(lojas_df, on='ID Loja').rename(columns={'E-mail' : 'E-mail do Cliente'})
display(novas_vendas2017_df.head())
```

Achar um dado específico na tabela:
    - duas formas:
        - loc - achar um dado conforme uma linha e uma coluna por meio de indice. Podendo transformar a linha em indice e achar por ela, assim como na coluna
        - iloc - sempre acha pela números (ex: dataframe.iloc[num_linha, num_coluna])
        
    - .set_index('argumento') - transforma a coluna argumento em índice 
    - .head(argumento) - pegar as primeiras linhas do dt ## Padrao 5 linhas se não colocar nada
    - .tail(argumento) - pegar as últimas linhas do dt ## Padrao 5 linhas se não colocar nada


```python
# pegando o dado utilizando loc
display(novas_vendas2017_df.loc[[2,3,6,8]])

# transformando a coluna id produto em índice
novas_vendas2017_df = novas_vendas2017_df.set_index('Data da Venda')
#print(novas_vendas2017_df.head())

# pegando o dado utilizando loc e indice
print(novas_vendas2017_df.loc['11/11/2017', 'ID Produto'].head())
```

```python
# pegando o dado utilizando iloc
display(novas_vendas2017_df.iloc[2, 10])
```

Alterar um dado específico no df
    - achar o dado que quer e atribui atualiza o dado

```python
# Substituindo um dado por outro
## loc:
novas_vendas2017_df.loc[novas_vendas2017_df['Numero da Venda'] == 102415, ['Quantidade Vendida']] = 69
display(novas_vendas2017_df)

## iloc:
```
