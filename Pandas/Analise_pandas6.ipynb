```python
import pandas as pd

# dando entrada nas tabelas que serÃ£o usadas
produtos_df = pd.read_csv(r'E:\Programas\Projetos-Python\Pandas\Contoso - Cadastro Produtos.csv', sep=';')
clientes_df = pd.read_csv(r'E:\Programas\Projetos-Python\Pandas\Contoso - Clientes.csv', sep=';')
lojas_df = pd.read_csv(r'E:\Programas\Projetos-Python\Pandas\Contoso - Lojas.csv', sep=';')
vendas2017_df = pd.read_csv(r'E:\Programas\Projetos-Python\Pandas\Contoso - Vendas  - 2017.csv', sep=';')

# filtrando as tabelas conforme as colunas que queremos
clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto', 'Nome da Marca']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]

# mesclando e renomeando colunas
vendas2017_df = vendas2017_df.merge(produtos_df, on='ID Produto')
vendas2017_df = vendas2017_df.merge(clientes_df, on='ID Cliente')
vendas2017_df = vendas2017_df.merge(lojas_df, on='ID Loja').rename(columns={'E-mail' : 'E-mail do Cliente'})
display(vendas2017_df)
```

Exportar um dataframe como um csv:
    - dataframe.to_csv('nome_do_arquivo.csv', sep=';')

```python
# alternado o tamanho do arquivo
vendas2017_df = vendas2017_df[:10]

# exportar um dataframe como um csv
vendas2017_df.to_csv(r'E:\Programas\Projetos-Python\Pandas\Novo Vendas - 2017.csv', sep=';')
```

Transformar um dicionÃ¡rio em um dataframe e exportar o mesmo para um csv:
    - pd.DataFrame.from_dict(dicionÃ¡rio, orient='argumento') - usar index se for inverter coluna e linha, deixar vazio se n quiser inverter
    - encoding='argumento' - argumento Ã© para fazer com que o arquivo leia corretamente as palavras da nossa linguagem


```python
# criando um dicionÃ¡rio
vendas_celulares = {'iphone' : [19023, 30495], 'A51' : [12093, 3214], 'Xiomi' : [123,4543], 'Galaxy69' : [1234, 5234]}

# passando o dicionÃ¡rio para um df
vendas_celulares_df = pd.DataFrame.from_dict(vendas_celulares, orient='index')
display(vendas_celulares_df)

# renomeando as colunas
vendas_celulares_df = vendas_celulares_df.rename(columns={0 : 'Vendas 2019', 1 : ' Vendas 2020'})
display(vendas_celulares_df)

# exportando para csv
vendas_celulares_df.to_csv(r'E:\Programas\Projetos-Python\Pandas\Vendas Celulares.csv', sep=';', encoding='latin1')
```
