```python
import pandas as pd


# pegando o link direto dos dados e dando entrada
url_cotacao = r'https://ptax.bcb.gov.br/ptax_internet/consultaBoletim.do?method=gerarCSVFechamentoMoedaNoPeriodo&ChkMoeda=61&DATAINI=30/06/2022&DATAFIM=30/12/2022'
cotacao_dolar_df = pd.read_csv(url_cotacao, sep=';')

# dando nome as colunas
cotacao_dolar_df.columns = ['Data', 'Codigo da Moeda', 'Tipo', 'Moeda', 'Compra', 'Venda', '1', '2']

# formatando a coluna data
cotacao_dolar_df['Data'] = pd.to_datetime(cotacao_dolar_df['Data'], format='%d%m%Y')

#cotacao_dolar_df['30062022'] = pd.to_datetime(cotacao_dolar_df['Data da Venda'], format='%d/%m/%Y')
display(cotacao_dolar_df.head(10))
```

    - Pegar o link que passa por uma requisição e tratar ele:
        - import requests:
            - requests.get(r'argumento').content - vai pegar o conteudo que esta naquele argumento(link)
        - import io:
            - io,StringIO(argumento1.decode('argumento2')) - vai transformar arquivo que está em texto e vai interpretar ele como um arquivo csv:
                - argumento 1 - variavel criada no passo acima
                - argumento 2 - definições do decode como forma de interpretar a lingua em que está escrito (no caso, português brasil)        

```python
import pandas as pd
import requests
import io


# pegando o link dos dados
url_cafe = r'http://portalweb.cooxupe.com.br:8080/portal/precohistoricocafe_1.jsp?d-3961599-e=2&6578706f7274=1'

# tratando os dados
conteudo_url_cafe = requests.get(url_cafe).content

# transformando de texto para dataframe
arquivo_cafe = io.StringIO(conteudo_url_cafe.decode('latin1'))
cotacao_cafe_df = pd.read_csv(arquivo_cafe, sep='\t')

# definindo a coluna ano como index
cotacao_cafe_df = cotacao_cafe_df.set_index('Ano')

# colocando em ordem conforme o index
cotacao_cafe_df = cotacao_cafe_df.sort_index()
display(cotacao_cafe_df)
```
